```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/akc-data-latest.csv')
df.rename(columns={'Unnamed: 0': 'Breed'}, inplace=True)
df.head()
```

```python
unique_values = set(df['group'].unique())
print(unique_values)
new_columns = [unique_values]

for col in unique_values:
    df[col] = 0
```

```python
df.head()
```

```python
for index, row in df.iterrows():
  group = row['group']
  df.loc[index, group] = 1

df.head()
```

```python
column_names = df.columns.tolist()
print(column_names)
```

```python
df = df[['Breed', 'min_height', 'max_height', 'min_weight', 'max_weight', 'min_expectancy', 'max_expectancy', 'grooming_frequency_value', 'shedding_value', 'energy_level_value', 'trainability_value', 'demeanor_value', 'Herding Group', 'Toy Group', 'Non-Sporting Group', 'Terrier Group', 'Foundation Stock Service', 'Hound Group', 'Working Group', 'Sporting Group']]
df.head()
```

```python
df_ind = df.set_index('Breed')
df_ind.head()
```

```python
from scipy.sparse import lil_matrix
import scipy.spatial.distance
```

```python
queries = ['Bernese Mountain Dog', 'Newfoundland', 'Samoyed']
for q in queries:
  print(q)
  dog_data = df_ind.loc[q]
  distances = scipy.spatial.distance.cdist(df_ind, [dog_data], metric = "euclidean")[:,0]
  query_distances = list(zip(df.index, distances))

  for dog, characteristics in sorted(query_distances, key = lambda x: x[1], reverse = False)[:10]:
    print(df['Breed'].iloc[dog], characteristics)

  print("\n")
```
